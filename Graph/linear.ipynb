```python
# Data operation through pandas-
# 1.print(df.head()) --> Displays the first 5 rows of the dataset. and Useful to quickly check the structure and sample data.
import pandas as pd 
df=pd.read_csv("ncr_ride_bookings.csv")  
print(df.head())
```

```python
# 2. print(df.tail(10)) --> Displays the last 10 rows of the dataset. and Helps in verifying if the data was read correctly till the end.
print(df.tail(10))
```

```python
# 3. print(df.describe()) --> Generates summary statistics for numerical columns:-count,mean,std,min,median,max. and Very useful for understanding data distribution.
import pandas as pd
df=pd.read_csv("ncr_ride_bookings.csv") 
print(df.describe())
```

```python
# 4. print(df.info()) --> Prints metadata about the DataFrame:- a.Number of rows and columns, b.Column names and data types (int, float, object, etc.), c.
#    Count of non-null values per column.
print(df.info())
```

```python
# 5. df.dropna() --> Removes rows with missing values (NaN).
df.dropna()
```

```python
# 6. df = df.fillna(0) --> Replaces all missing values (NaN) in the dataset with 0. and This ensures there are no null values left in df.
df = df.fillna(0)
```

```python
# 7. df.isnull().sum() --> Shows the count of missing values per column. and Helps identify which columns have missing data.

df.isnull().sum()
```

```python
# 8. df.isnull().any() --> Returns True/False for each column, showing whether it contains any missing values.
df.isnull().any()
```

```python
# 9. df.isnull().sum().sum() --> Gives the total number of missing values in the whole DataFrame.
df.isnull().sum().sum()
```

```python
# 1. Simple Line Graph
import matplotlib.pyplot as plt
x = [10, 20, 30, 40]
y = [70, 80, 90, 100]

plt.plot(x, y, marker='o')    # marker='o' adds dots at points
plt.title("Simple Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# 2. scatter
import pandas as pd
import matplotlib.pyplot as plt

study_hours = [10, 20, 30, 50, 60, 70, 80]
grades = [1, 2, 3, 4, 5, 6, 7]

df = pd.DataFrame({
    "Study Hours": study_hours,
    "Grades": grades
})

plt.scatter(df["Study Hours"], df["Grades"], color='blue', marker='o')
plt.xlabel("Study Hours")
plt.ylabel("Grades")
plt.title("Study Hours vs Grades (Scatter Plot)")
plt.grid(True)
plt.show()
```

```python
# 3. Histogram
plt.hist(df["Study Hours"], bins=5, color='blue', edgecolor='black')
plt.xlabel("Study Hours")
plt.ylabel("Frequency")
plt.title("Study Hours Histogram")
plt.show()
```

```python
#4. bar graph 
import matplotlib.pyplot as plt

# Data
x = [10, 20, 30, 40]   # Study Hours
y = [70, 80, 90, 100]  # Grades

# Bar Graph
plt.bar(x, y, color='blue', edgecolor='black')
plt.xlabel("Study Hours")
plt.ylabel("Grades")
plt.title("Study Hours vs Grades (Bar Chart)")
plt.show()
```

```python
# 5. Pie Chart
plt.pie(df["Study Hours"], labels=df["Grades"], autopct="%1.1f%%", startangle=90)
plt.title("Study Hours Distribution (Pie Chart)")
plt.show()
```
